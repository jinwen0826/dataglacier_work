```python
import matplotlib.pyplot as plt
import seaborn as sns
df=bank_full 
# Draw Plot
plt.figure(figsize=(16,10), dpi= 80)
sns.kdeplot(df.loc[df['duration'] <= 5000, "duration"], shade=True, color="g",label='Duration')

# Decoration
plt.title('Density Plot of duration', fontsize=22)
plt.legend()
plt.show()
```
